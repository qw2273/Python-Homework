# define wrLines() and wrChars() functions
- note that 'tobe' has newline chars(\n), but there are none in 'tobeLines'
- both functions should first remove the file named by tp if it exists, to get a fresh start.
- os.remove will throw an error if the file doesn't exist,
so either catch it or verify the file exists before trying to delete it
    - wrLines writes the contents of 'tobeLines' to a file, using the fd.writelines() method. 
you must add a '\n' explicitly to each line. if you are on windows, use '\n' as your newline character, so the counts come out correctly
    - wrChars writes the contents of 'tobe', one character at a time, using fd.write() method
- after the file write is complete, both functions should get the file byte size,
using os.stat, and return it.

````python
import os

# get a tmp file pathname. all functions below will use it

import tempfile

# does not create the file, just makes a pathname
tp = tempfile.NamedTemporaryFile().name
````

````python
def wrLines(temp, to_write):
    # check if file exists - if so remove 
    try:
        os.remove(temp) 
    except OSError:
        pass
    with open(temp, 'w') as fd:
        for e in to_write:
            
            fd.write(e + '\n')
        
    return os.stat(temp).st_size
          
````

````python
wrLines(tp, tobeLines)
````

````python
def wrChars(temp, to_write):
    try:
        os.remove(temp) 
    except OSError:
        pass
    with open(temp, 'w') as fd:
        for s in list(to_write):
            fd.write(s)
    return os.stat(temp).st_size
          
````

# define rdAll()
- read the tp file in one fd.read() call (eagar approach)
- return the number of characters read
- if optional search arg supplied, return length of chars read, and index of where search string is found

````python
def rdAll(temp, search = None):
    with open(temp, 'r') as fd:
        x = fd.read()
        #print(x)
    if search is None:
        return len(x)
    else:
        return [ len(x),  x.index(search)]
    
````

# define rdLines()
- read the tp file by using fd.readlines() (eager approach)
- with no arg, return the number of characters read
- with optinal search arg, return the index of the search string in a list

````python
def rdLines(temp, search = None):
    with open(temp, 'r') as fd:
        x = fd.readlines()
    s = "".join(x)
    if search is None:
        return len(s)
    else:
        return  [ s.index(search) ]
        
````

# define rdChars
- read the tp file one character at a time using fd.read(1) (lazy approach)
- fd.read(1) will return an empty string when the file is exhausted
- with no arg, return the number of characters read
- with optinal search arg, return the index of the search string in a list

````python
def rdChars(temp, search = None):
    with open(temp, 'r') as fd:
        res = []
        while True:
            s = fd.read(1)
            if len(s) == 0 :
                break; 
            else:
                res.append(s)
    l = "".join(res)
    if search is not None:
        return [l.index(search)]
    else:
        return len(l)
````

````python
def readHamlet():
    with urllib.request.urlopen(hamleturl) as ef:
        return [l.decode('utf-8') for l in ef]
````

````python
def hamlet():
    # subtract all speeches and each speech is an element in the list
    s_split = re.split('A NAME=speech', ''.join(readHamlet()) )
    
    # for each speech, subtract the name and store it in the dictionary
    d = dict()
    for s in s_split:
        if re.findall('b>[a-zA-Z]+\s*[a-zA-Z]+</b',s) != []:
            name = re.findall('b>[a-zA-Z]+\s*[a-zA-Z]+</b',s)[0][2:-3]
            if name not in d.keys():
                d[name] = 1
            else:
                d[name] += 1
    
    return(len(readHamlet()),len(s_split)-1, d)
    
````

````python
from datetime import datetime
import pandas as pd
````

````python
# with the arg defaults shown, this will 
# calculate the money accumulated after 
# 31 days, with weekly deposits of $10, 
# at a 10% interest rate compounded
# on each business day

def savings(start = '2018-03-01', end = '2018-03-31',\
            depositFreq='W', depositAmount=10., 
            compoundFreq='B', interestRate=.1 ):
    # your code here
    money = pd.DataFrame(index = pd.date_range(start,end))
    money['day'] = [d.strftime('%A') for d in pd.date_range(start,end)]
    money['deposits'] = pd.Series(depositAmount, index = pd.date_range(start,end,freq = depositFreq))
    money['compound'] = pd.Series(interestRate, index =pd.date_range(start,end,freq = compoundFreq))
    

    intser = pd.Series(0.0,index =pd.date_range(start,end))
    balser = pd.Series(0.0, index =pd.date_range(start,end))
    money = money.fillna(0)
    
    balance = 0.0
    interest = 0.0
        
    for j,t in enumerate(money.itertuples()):
        # calculate interest and new balance
        balance += t.deposits
        interest = balance * t.compound
        balance += interest
        intser[j] = interest
        balser[j] = balance
        # update the series objects
    money['interest'] = intser
    money['balance'] = balser
    

    return money
    
````
